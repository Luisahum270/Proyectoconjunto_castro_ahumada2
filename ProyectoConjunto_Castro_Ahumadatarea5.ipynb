Para encontrar la función de densidad $g(x)$, que es la función de densidad de la distribución de probabilidad $U=F(X)$, donde $F(X)$ es la función de distribución acumulada de $X$, podemos utilizar la siguiente fórmula:

$$g(x) = f(x) \left|\frac{d}{dx}F(x)\right|$$

Para nuestra distribución, la función de distribución acumulada es:

$$F(x) = \int_0^x f(t) dt = 5x^2 - 6x^3 + 2x^4$$

Entonces, la función de densidad $g(x)$ es:

\begin{align*}
g(x) &= f(x) \left|\frac{d}{dx}F(x)\right| \
&= 30(x^2 - 2x^3 + x^4) \left|\frac{d}{dx}(5x^2 - 6x^3 + 2x^4)\right| \
&= 30(x^2 - 2x^3 + x^4) \left|(10x - 18x^2 + 8x^3)\right| \
&= 30(x^2 - 2x^3 + x^4) (10x - 18x^2 + 8x^3) \
&= 300x^5 - 840x^6 + 720x^7 - 180x^8
\end{align*}

Entonces, la función de densidad $g(x)$ es $g(x) = 300x^5 - 840x^6 + 720x^7 - 180x^8$.

La distribución acumulada $G(x)$ se encuentra integrando la función de densidad de probabilidad $g(x)$ desde $0$ hasta $x$:

$$G(x) = \int_{0}^{x} g(t) dt = \int_{0}^{x} (300t^5 - 840t^6 + 720t^7 - 180t^8) dt $$

Calculando la integral:

$$G(x) = 60x^6 - 140x^7 + 120x^8 - 45x^9 $$

La función inversa $G^{-1}(u)$ se encuentra despejando $x$ de la ecuación anterior:

$$u = G(x) = 60x^6 - 140x^7 + 120x^8 - 45x^9 $$

Resolviendo para $x$, obtenemos:

$$x = G^{-1}(u) = \frac{1}{9} \ln \left(\frac{20u}{3} + \frac{\sqrt{3}\sqrt{135u^2 - 140u + 16}}{3}\right)$$

El método de la transformada inversa es una técnica que nos permite generar variables aleatorias que siguen una distribución de probabilidad específica a partir de una distribución uniforme en el intervalo [0,1].

Primero, necesitamos calcular la distribución acumulada de la función de densidad g(x). La distribución acumulada se define como:

$$G(x) = \int_{-\infty}^{x} g(t) dt$$

Para nuestra función de densidad g(x) = a sin(πx), la distribución acumulada se puede calcular de la siguiente manera:

$$G(x) = \int_{-\infty}^{x} g(t) dt = \int_{-\infty}^{x} a \sin(\pi t) dt = \frac{-a}{\pi} \cos(\pi x) + \frac{a}{\pi}$$

Ahora, necesitamos invertir la distribución acumulada G(x) para obtener la función inversa G⁻¹(u) que se usa para generar variables aleatorias. La función inversa se puede calcular de la siguiente manera:

$$u = G(x) = \frac{-a}{\pi} \cos(\pi x) + \frac{a}{\pi}$$

Despejando x, obtenemos la función inversa:

$$x = G^{-1}(u) = \frac{1}{\pi} \cos^{-1}(\frac{\pi u - 1}{a})$$

```python
# Función de aceptación y rechazo usando compresión de listas
def Acep_rechazo(R2:'Variables distruidas U~U(0,1)',
                 R1:'Variables distribuidas como g(x)',
                 f:'función objetivo a generar',
                 t:'función que mayora a f'):
#     R1 = np.random.rand(N)
    f_x = f(R1)
    t_x = t(R1)
    condition = R2*t_x <=f_x
    [plt.plot(R1[i],R2[i]*t_x[i],'ob') if condition[i] else plt.plot(R1[i],R2[i]*t_x[i],'o') \
         for i in range(len(R1))]
                
    plt.show()
```

```python
# Ilustración del método de aceptación y rechazo cuando se toma t(x) constante

# Función objetivo 
f = lambda x: 30 * (x**2 - 2*x**3 + x**4)
# Máximo de la función f
max_f = f(optimize.fmin(lambda x:-f(x), 0, disp=False))
# Función t -> Función constante
t = lambda x: max_f * np.sin(np.pi * x)

# Rango donde se graficará las funciones
x = np.arange(0, 1, 0.01)
print('El máximo de f es:',max_f)

# Gráficas de las funciones
plt.plot(x,f(x),label='f(x)')
plt.plot(x,t(x),label='t(x)')
plt.legend()

# Validación del método
N = 500 # número de puntos a simular
# Como estoy tomando t(x) constante solo es necesario generar valores aleatorios U~(0,1)
R2 = np.random.rand(N)
R1 = np.random.rand(N)

# Prueba de función creada
Acep_rechazo(R2, R1, f, t)
```

```python
# Función de aceptación y rechazo usando compresión de listas
def Acep_rechazo3(R2:'Variables distruidas U~U(0,1)',
                 R1:'Variables distribuidas como g(x)',
                 f:'función objetivo a generar',
                 t:'función que mayora a f'):
    f_x = f(R1)
    t_x = t(R1)
    condition = R2*t_x <=f_x
    accepted_points = np.sum(condition)
    total_points = len(R1)
    print("Porcentaje de puntos aceptados:", accepted_points/total_points)
    [plt.plot(R1[i],R2[i]*t_x[i],'ob') if condition[i] else plt.plot(R1[i],R2[i]*t_x[i],'or') \
         for i in range(len(R1))]
                
    plt.show()

# Función objetivo 
f = lambda x: 30 * (x**2 - 2*x**3 + x**4)
# Máximo de la función f
max_f = f(optimize.fmin(lambda x:-f(x), 0, disp=False))

# 1. Usando t(x) constante
# Función t -> Función constante
t_const = lambda x: max_f * np.ones_like(x)

# Prueba de función creada
N = 10000 # número de puntos a simular
R2 = np.random.rand(N)
R1 = np.random.rand(N)
Acep_rechazo(R2, R1, f, t_const)


# 2. Usando t(x) un pulso senoidal
# Función t -> Pulso senoidal
t_sin = lambda x: max_f * np.sin(np.pi * x)

# Prueba de función creada
N = 10000 # número de puntos a simular
R2 = np.random.rand(N)
R1 = np.random.rand(N)
Acep_rechazo3(R2, R1, f, t_sin)
```
